```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/updated_engine_data (1).csv')
```

```python
data.info()
```

```python
data.head()
```

```python
#derive new features from existing features
data['Engine_power'] = data['Engine rpm'] * data['Lub oil pressure']
```

```python
data.describe()
```

```python
data['Temperature_difference'] = data['Coolant temp'] - data['lub oil temp']
```

```python
data.describe().T
```

```python
data.drop(['Engine_power'], axis=1, inplace=True)
```

```python
data['Fuel Type'] = data['Fuel Type'].map({'Diesel': 0.0, 'Petrol': 1.0})
```

```python
data['Fuel Type']
```

```python
data['Fuel Type'] = data['Fuel Type'].astype(float)
```

```python
data.info()
```

```python
#use the GBM model to predict the engine status
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = data.drop(['Engine Condition'], axis=1)
y = data['Engine Condition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42, max_features='sqrt',min_samples_leaf=5,min_samples_split=2,subsample=0.8)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)

```

```python
 #give f1 score report with precision and recall
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pickle
import numpy as np
from datetime import datetime, timedelta

# Load the trained model
with open('/content/hhmodel.pkl', 'rb') as file:
    model = pickle.load(file)

# Define a single test instance (replace these values with actual test data)
test_instance = np.array([520,	2.961746,	6.553147,	1.064347,	77.752266,	79.645777,	0.0,	8.337864,	4.335690,	1168.208944,	1.893511]).reshape(1, -1)

# Define threshold probability and other constants
threshold = 0.6  # Probability threshold for maintenance recommendation
max_days_to_maintenance = 30  # Maximum possible days to maintenance when probability is 1.0

# Perform prediction
probability = model.predict_proba(test_instance)[0, 1]  # Probability for class 1
prediction = model.predict(test_instance)[0]  # Predicted class (0 or 1)

# Current date as a starting point
current_date = datetime.now()

# Calculate approximate days to maintenance
if probability > threshold:
    days_to_maintenance = int(max_days_to_maintenance * probability)
    maintenance_date = current_date + timedelta(days=days_to_maintenance)
    print(f"Maintenance is predicted to be needed around {maintenance_date.strftime('%Y-%m-%d')} (in approximately {days_to_maintenance} days).")
else:
    print("Engine is in good condition; no immediate maintenance needed.")

# Detailed output
print(f"\nPredicted Class: {prediction}")
print(f"Probability for Maintenance Need: {probability:.2%}")
print(f"Threshold: {threshold:.2%}")
```
